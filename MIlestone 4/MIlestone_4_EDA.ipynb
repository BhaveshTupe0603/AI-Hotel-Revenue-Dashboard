```python
# ----------------------------------------
# üìä Exploratory Data Analysis (EDA)
# ----------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hotel_bookings_cleaned.csv")

# Create necessary calculated columns
df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df['revenue'] = df['adr'] * df['total_nights']

# Display basic info
print("Dataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())

# Display summary
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['adr'], bins=40, edgecolor='black')
plt.title("Distribution of ADR (Average Daily Rate)")
plt.xlabel("ADR Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,5))
df.groupby('hotel')['revenue'].sum().plot(kind='bar', color=["steelblue","orange"])
plt.title("Revenue by Hotel Type")
plt.xlabel("Hotel")
plt.ylabel("Total Revenue")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby('market_segment')['revenue'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Revenue by Market Segment")
plt.xlabel("Booking Source")
plt.ylabel("Total Revenue")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df[df['is_canceled'] == 1].groupby('distribution_channel').size().plot(kind='bar', color='crimson')
plt.title("Cancellations by Distribution Channel")
plt.xlabel("Distribution Channel")
plt.ylabel("Count of Cancellations")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby('customer_type')['total_nights'].sum().plot(kind='bar', color='green')
plt.title("Total Nights Stayed by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Total Nights")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby('country')['revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Revenue Generating Countries")
plt.xlabel("Country")
plt.ylabel("Revenue")
plt.show()
```
